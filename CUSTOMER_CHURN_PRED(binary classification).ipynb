```python
import numpy as np # linear algebra
import pandas as pd
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/JATINUPRETI10/cc_churn/refs/heads/main/Churn_Modelling.csv')
```

```python
df.head()
```

# removing unnesscary col

```python
df.drop(columns=['RowNumber','CustomerId','Surname'],inplace=True)
```

```python
df.head()
```

```python
df.value_counts('Geography')
```

```python
df.value_counts('Gender')
```

```python
df=pd.get_dummies(df,columns=['Geography','Gender'],drop_first=True)
```

```python
df.head()
```

```python
df=df.astype(int)
```

```python
df.head()
```

# spliting and training

```python
x=df.drop(columns=['Exited'])
y=df['Exited'].values
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

# fit transforming dataset

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

# deep learing started

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(3,activation='sigmoid',input_dim=11))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy')
```

```python
model.fit(x_train,y_train,epochs=50)
```

```python
model.layers[1].get_weights()
```

```python
y_log=model.predict(x_test)
```

```python
y_pred=np.where(y_log>0.5,1,0)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
